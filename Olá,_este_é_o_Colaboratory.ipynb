```
def master_t (a):
    print("Medidas de Centralidade e Dispersão para ",a)
    print("media: ",master[a].mean())
    print("desvio padrão: ",master[a].std())
    print("mediana: ",master[a].median())
    print("variancia: ",master[a].var())
    print("moda:\n",master[a].mode())
    print("boxplot para ",a)
    master.boxplot(column=a)
```
